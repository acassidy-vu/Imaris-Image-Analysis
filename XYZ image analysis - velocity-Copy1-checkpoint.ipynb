```python
import csv
    
    
#___________________________X velocity____________________________    
x_velocity = []
x_velocity_f=[]
with open(r'C:\Users\annao\\Documents\Zanic Rotation\LLCPK1 GFP-EB1  257_iSIM-488nm-525-50 x velocity_Detailed.csv','r') as x_vel:
    header=x_vel.readline()
    heade=x_vel.readline()
    head=x_vel.readline()
    hea=x_vel.readline()
    he=x_vel.readline()
    h=x_vel.readline()
    for line in x_vel:
        X_V=line.rstrip().split(',')
        #xv=float(xvel[0])
        x_velocity.append(X_V[0])
x_velocity.remove('Velocity X')
try:
    while True:
        x_velocity.remove('')
except ValueError:
    pass

for values in x_velocity:
    v=float(values)
    if v<=0:
        vv=v*-1
        x_velocity_f.append(vv)
    else:
        x_velocity_f.append(v) 
        
            
#print(x_velocity_f)
    
    
#___________________________Y velocity____________________________    
y_velocity = []
y_velocity_f=[]
with open(r'C:\Users\annao\\Documents\Zanic Rotation\LLCPK1 GFP-EB1  257_iSIM-488nm-525-50 y velocity_Detailed.csv','r') as y_vel:
    header=y_vel.readline()
    heade=y_vel.readline()
    head=y_vel.readline()
    hea=y_vel.readline()
    he=y_vel.readline()
    h=y_vel.readline()
    for line in y_vel:
        Y_V=line.rstrip().split(',')
        #xv=float(xvel[0])
        y_velocity.append(Y_V[0])
y_velocity.remove('Velocity Y')
try:
    while True:
        y_velocity.remove('')
except ValueError:
    pass

for values in y_velocity:
    v=float(values)
    if v<=0:
        vv=v*-1
        y_velocity_f.append(vv)
    else:
        y_velocity_f.append(v)     
    
#___________________________Z velocity____________________________    
z_velocity = []
z_velocity_f=[]
with open(r'C:\Users\annao\\Documents\Zanic Rotation\LLCPK1 GFP-EB1  257_iSIM-488nm-525-50 z velocity_Detailed.csv','r') as z_vel:
    header=z_vel.readline()
    heade=z_vel.readline()
    head=z_vel.readline()
    hea=z_vel.readline()
    he=z_vel.readline()
    h=z_vel.readline()
    for line in z_vel:
        Z_V=line.rstrip().split(',')
        #xv=float(xvel[0])
        z_velocity.append(Z_V[0])
z_velocity.remove('Velocity Z')
try:
    while True:
        z_velocity.remove('')
except ValueError:
    pass

for values in z_velocity:
    v=float(values)
    if v<=0:
        vv=v*-1
        z_velocity_f.append(vv)
    else:
        z_velocity_f.append(v) 
    
    
#__________________Graphs__________________
import matplotlib.pyplot as plt
import numpy as np


x7=x_velocity_f
y7=y_velocity_f
z7=z_velocity_f

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(x7,y7,z7,s=4, c='b', alpha=0.09)
ax.set_title('EB1 Comet Velocity in LLCPK1 Cells', fontsize=25)
ax.set_xlabel('X Velocity (um/s)', fontsize=15)
ax.set_ylabel('Y Velocity (um/s)', fontsize=15)
ax.set_zlabel('Z Velocity (um/s)', fontsize=15)

#xy_fig = plt.xyfig()
#plt.scatter(x4,y4,s=8, c='r', alpha=0.2)
#plt.title('EB1 Comet Velocity in LLCPK1 Cells')
#plt.xlabel('X Velocity')
#plt.ylabel('Y Velocity')
#plt.zlabel('Z Velocity')
ax.figure.savefig('3D velocity distribution cell 257.jpg', dpi=300, bbox_inches='tight')
```

```python
x_velocity_avg=[]
y_velocity_avg=[]
z_velocity_avg=[]

import csv
with open (r'C:\Users\annao\Desktop\Velocity.csv','r') as avg_velocity:
    header=avg_velocity.readline()
    for line in avg_velocity:
        AV=line.rstrip().split(',')
        x_velocity_avg.append(float(AV[0]))
        y_velocity_avg.append(float(AV[1]))
        z_velocity_avg.append(float(AV[2]))

#print(x_velocity_avg)
#print(y_velocity_avg)
#print(z_velocity_avg)

x=x_velocity_avg
y=y_velocity_avg
z=z_velocity_avg

import seaborn as sns
green_diamond = dict(markerfacecolor='b', marker='.', alpha=0.08)
red= dict(color='r')
blue= dict(color='b')
labels=list('XYZ')
data = [x, y, z]
fig, ax = plt.subplots()
ax.set_title('Distribution of EB1 Comet Velocity', fontsize=20)
ax.set_xlabel('Dimension', fontsize=15)
ax.set_ylabel('Velocity (um/s)', fontsize=15)
AX= ax.boxplot(data, showmeans=True, labels=labels, flierprops=green_diamond, 
            medianprops=red, meanprops=blue)

#for i in [1,2,3]:
    #y = AX4.distribution[titanic.dimension==i].dropna()
    #ax4.scatter(data, 'r.', alpha=0.2)
    


medians = [item.get_ydata()[0] for item in AX['medians']]
means = [item.get_ydata()[0] for item in AX['means']]
q1 = [round(min(item.get_ydata()), 1) for item in AX['boxes']]
q3 = [round(max(item.get_ydata()), 1) for item in AX['boxes']]

print(f'\033[1mX Velocity: \033[0m\n'
        f'Median: {medians[0]}\n'
      f'Mean: {means[0]}\n'
     f'Q1: {q1[0]}\n'
     f'Q3: {q3[0]}\n')
print(f'\033[1mY Velocity: \033[0m\n'
    f'Median: {medians[1]}\n'
      f'Mean: {means[1]}\n'
     f'Q1: {q1[1]}\n'
     f'Q3: {q3[1]}\n')
print(f'\033[1mZ Velocity: \033[0m \n'
    f'Median: {medians[2]}\n'
      f'Mean: {means[2]}\n'
     f'Q1: {q1[2]}\n'
     f'Q3: {q3[2]}\n')

from scipy import stats
stat=stats.kruskal(x, y, z)
print(stat)
print(f'The p value is {round(stat[1],2)}')

plt.savefig('Boxplot of Average Velocity Distribution.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
